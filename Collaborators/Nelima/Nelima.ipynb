```python
# Read the csv files

gsv = pd.read_csv('../../Datasets/bom.movie_gross.csv')
gsv.head()
```

## Data Cleaning

```python
#Drop duplicate rows
gsv = gsv.drop_duplicates()
gsv 
```

```python
#Changing the data type in values in column foreign gross to a float
gsv['foreign_gross'] = gsv['foreign_gross'].astype(str).str.replace(',', '').astype(float)
print(gsv['foreign_gross'].dtype)
```

```python
gsv.isna().sum()
```

```python
# Remove rows with any NaN values
cleaned_gsv = gsv.dropna(axis=0, how='any')

# Remove columns with any NaN values
cleaned_gsv = cleaned_gsv.dropna(axis=1, how='any') 

# Select numeric columns
numeric_columns = cleaned_gsv.select_dtypes(include=['float64', 'int64']).columns 

# Calculate the mean for each numeric column
column_means = cleaned_gsv[numeric_columns].mean() 
# Fill NaN values with the mean of each column
cleaned_gsv = cleaned_gsv.fillna(column_means) 
cleaned_gsv
```

```python
# Calculate the total gross for each movie
gsv['total_gross'] = gsv['domestic_gross'] + gsv['foreign_gross']

# Sort DataFrame by total gross in descending order to identify the most profitable movies
most_profitable_movies = gsv.sort_values(by='total_gross', ascending=False)

# Optionally, display the dates and studios where the most profitable movies were shot
most_profitable_movies_with_details = most_profitable_movies[['studio', 'total_gross', 'year']]

# Display the top 10 most profitable movies along with their shooting dates and studios
most_profitable_movies_with_details.head(10)
```

```python
# add one column of total revenues 
cleaned_gsv["total_gross"]=cleaned_gsv["domestic_gross"] + cleaned_gsv["foreign_gross"]
```

### Hypothesis Testing (Anova)

We will set a hypotheses test to determine if there are statistically significant differences between the means of different revenues for the studios.To check if there are significant differences amongst the studios we calculate the F-statistic and p-values using the Anova test

##### Null Hypothesis (H₀)
##### H₀: The means of total revenues for all studios are equal.
##### Alternate Hypothesis (H₁)
##### H₁: At least one studio has a mean total revenue that is different from the others.

```python

from statsmodels.formula.api import ols
import statsmodels.api as sm
from statsmodels.stats.anova import anova_lm

# Filter the data for the top 20 studios
top_20_studios = cleaned_gsv[cleaned_gsv['studio'].isin(Total_Revenue_by_studio.head(20)['studio'])]

# Perform ANOVA
# Example: Comparing Total Revenue across different studios
model = ols('Total_Revenue ~ C(studio)', data=top_20_studios).fit()
anova_results = anova_lm(model)

# Display ANOVA table
print(anova_results)
```

The ANOVA results indicate that there are statistically significant differences in the total revenues between different studios. This means that not all studios generate the same revenue, and some studios are likely to have higher revenues than others.
F-value of 14.38 indicates a significant difference between the studios’ total revenues.
The p-value of 8.44-42 is much lower than 0.05, suggesting the differences between the mean revenues of the studios are statistically significant and therefore reject the null hypothesis
